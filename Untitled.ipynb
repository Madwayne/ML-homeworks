```python
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
goods_specs = pd.read_csv('Спецификации_лекарственных_средств.csv', sep=';', engine='python')
goods_specs
```

```python
goods_descr = pd.read_json('Спецификации_лекарственных_средств_-_многострочные_поля.json', encoding = 'Windows-1251')
```

```python
goods_descr
```

```python
goods = pd.merge(goods_specs, goods_descr, on='ID')
goods = goods[['ID', 'Сокращённое наименование', 'Способы применения', 'Показания', 'Противопоказания', 'Описание']].set_index('ID')
goods
```

```python
from ftfy import fix_text
```

```python
def extract_numbers_phrases(text):
    numbers_phrases = []
    if text:
        norm_text = fix_text(text, normalization='NFKC').replace('\n', ' ').replace('\n', ' ')
        for i in range(len(norm_text.split(' '))):
            if norm_text.split(' ')[i].isdigit():
                numbers_phrases.append(' '.join(norm_text.split(' ')[i-1:i+2]))
    return numbers_phrases
```

```python
goods['numbers_phrases'] = goods['Способы применения'].apply(extract_numbers_phrases) + \
                           goods['Показания'].apply(extract_numbers_phrases) + \
                           goods['Противопоказания'].apply(extract_numbers_phrases) + \
                           goods['Описание'].apply(extract_numbers_phrases)
```

```python
goods['target_numbers_phrases'] = goods['numbers_phrases'].apply(lambda x: [y for y in x if y.find(' лет') != -1 or \
                                                                            y.find(' мес') != -1 or \
                                                                            y.find(' год') != -1])
```

```python
goods['numbers_count'] = goods['target_numbers_phrases'].apply(lambda x: len(x))
```

```python
print('Товаров с хотя бы одним числовым значением в спецификации: ', len(goods[goods['target_numbers_phrases'].map(len) > 0]))
print('Товаров, ровно с одним числовым значением в спецификации: ', len(goods[goods['target_numbers_phrases'].map(len) == 1]))
```
